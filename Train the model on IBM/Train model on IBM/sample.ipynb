```python
sns.countplot(data=data, x = 'Dataset' )
LD,NLD=data[ 'Dataset' ] . value_counts( )
print("liver disease patinets: " , LD)
print("Non-1iver disease patinets: ",NLD)
```

```python
x=data.iloc[:,0:-1]
y=data. iloc[:,-1]
```

```python
from sklearn. model_selection import train_test_split
xtrain, xtest,ytrain, ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.svm import SVC
S=SVC()
S.fit(xtrain,ytrain)
t1=S.predict(xtest)
accuracy_score(t1,ytest)

```

```python
from sklearn.ensemble import RandomForestClassifier
R=RandomForestClassifier()
R.fit(xtrain,ytrain)
t2=R.predict(xtest)
accuracy_score(t2,ytest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
K=KNeighborsClassifier()
K.fit(xtrain,ytrain)
t3=K.predict(xtest)
accuracy_score(t3,ytest)
```
